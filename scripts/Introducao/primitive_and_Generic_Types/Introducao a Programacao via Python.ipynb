## Set (conjunto)

Um conjunto somente aceita valores únicos (não repetidos)

```python
meuConjunto = set()

meuConjunto.add(1)

meuConjunto.add(2)

meuConjunto.add(3)

meuConjunto.add("Coxinha")

meuConjunto.add("Abóbora")

meuConjunto.add("laranja")

meuConjunto.remove("laranja")

meuConjunto
```

```python
outroConjunto = set()

outroConjunto.add("Abóbora")
outroConjunto.add("Coxinha")
outroConjunto.add("Maracujá")

interseccao = meuConjunto.intersection(outroConjunto)
conjuntoDiferenca = meuConjunto.difference(outroConjunto)

print(type(interseccao), interseccao)

# Notar que a string "Maracujá" não faz parte destae conjunto diferença,
# pois este conjunto diferença é sempre tomado como
# base o conjunto da esquerda da operação

print(type(conjuntoDiferenca), conjuntoDiferenca, )

# Para considerar a diferença de ambos conjuntos, devemos utilizar a lógica de
# diferença simétrica
# neste caso, notar que a string "Maracujá" é retornada pela função.
conjuntoDiferencaSimetrica = meuConjunto.symmetric_difference(outroConjunto)

print(type(conjuntoDiferencaSimetrica), conjuntoDiferencaSimetrica)
```

## Álgebra Linear
   *   Matriz identidade
   *   Matriz Inversa
   *   Transposição de Matriz

```python
# Extraindo a Matriz identidade

Identidade = np.eye(matrix.shape[0])

# Produto Escalar
produtoEscalar_MI = matrix.dot(Identidade)
print(produtoEscalar_MI)

produtoEscalar_IM = Identidade.dot(matrix)

# Verificação de Permutação
PermutacaoEhIgual = np.all(produtoEscalar_MI == produtoEscalar_IM)

print(f"\n A permutação altera o produto? {~PermutacaoEhIgual}")
```

```python
# Matriz Inversa (se houver)

a = np.array([[1., 2.], [3., 4.]])
ainv = np.linalg.inv(a)
condicao = np.allclose(np.dot(a, ainv), np.eye(2))

print("Matriz original: ", matrix)

print(ainv)

print(f"Matriz 'A' vezes matriz 'Ainv' == matriz Identidade de A? \n\t Resposta: {condicao}")

```

```python
# Transposta de matrizes

A = np.array([[1,2,3], [4,5,6]])

print("Matriz original: \n", A, end=END)

print("Matriz transposta: \n", A.T, end=END)
```

### Sistema de Equações determinadas

${
f(x)=
\begin{cases}
x0 + 2 * x1 = 1 ,  \hspace{35pt} \text{eq 1 }\\
3 * x0 + 5 * x1 = 2 ,  \hspace{20pt} \text{eq 2 }
\end{cases}
}$

```python
# Equações determinadas
eq1 = [1, 2]
eq2 = [3, 5]

ladoEsquerdoDaIgualdade =  np.array([eq1,
                                     eq2])

ladoDireitoDaIgualdade = np.array([1, 2])
Resultado = np.linalg.solve(ladoEsquerdoDaIgualdade, ladoDireitoDaIgualdade)
x0, x1 = Resultado
print(f"Valores: \n\t x0 = {round(x0)} \n\t x1 = {round(x1)}")
```

## Equações sobre-determinadas (há mais equações do que variáveis)

Este tipo de situação envolve casos, por exemplo, em que buscamos traçar uma reta entre nossos pontos amostrais.

### Exemplo de caso:

$y = mx + c$

$x = [0, 1, 2, 3]$ => pontos amostrais da variável independente

$y = [-1, 0.2, 0.9, 2.1]$ => pontos amostrais da variável dependente





```python
x = np.array([0, 1, 2, 3])
y = np.array([-1, 0.2, 0.9, 2.1])

# dado que nossa equação f(x) aceita um coeficiente linear (i.e., o "c" do (fx)), devemos adicionar
# uma coluna unária (cheia de 1s).

A = np.vstack([x, np.ones(len(x))]).T

m, c = np.linalg.lstsq(A, y, rcond=None)[0]

print(f'Resposta: a equação linear da função é igual a "y = {round(m,2)}*x + {round(c, 2)}"')
```

## Funções recursivas

```python
def fibonaccy(n: int = 4):
  sequenciaFibonaccy = []
  penultimo=1
  ultimo=1

  if (n==1):
    sequenciaFibonaccy.append(penultimo)

  elif (n==2):
    sequenciaFibonaccy.append(penultimo)
    sequenciaFibonaccy.append(ultimo)

  else:
    sequenciaFibonaccy.append(penultimo)
    sequenciaFibonaccy.append(ultimo)


    while n >= 3:
      resultado = ultimo + penultimo
      penultimo = ultimo
      ultimo = resultado
      n -= 1
      sequenciaFibonaccy.append(resultado)

  for i in sequenciaFibonaccy:
    print(i)

  return sequenciaFibonaccy[-1]



print(fibonaccy(10))
```

```python
# versão recursiva

def fibonaccyRecursiva(n: int = 4, sequenciaFibonaccy = []):

  if (n==1):
    return 1

  elif (n==2):
    return 1

  else:
    return fibonaccyRecursiva(n -1) + fibonaccyRecursiva(n-2)

print(fibonaccyRecursiva(10))
```

# Tipos de Objeto Customizados

## Class (classes)

```python
class Pessoa:
  def __init__(self, nome, idade) -> None:
    self.nome = nome
    self.idade = idade

  def envelhecer(self) -> None:
    """Envelhece a pessoa em um ano.

    """
    self.idade = self.idade + 1


  def falar(self, x: str = None) -> str:
    """Fala"""

    if (x == None):
      x = f"Olá. Meu nome é {self.nome}. Eu tenho {self.idade} anos"
    return str(x)

  def morrer(self) -> None:
    """Deleta instância de Pessoa"""

    del self
```

```python
p1 = Pessoa("Philipe", 33)

print(p1.falar())
```

```python

# Vamos envelhecer a pessoa
print(f"Idade de p1 => {p1.idade}")

p1.envelhecer()

# Verificando idade de P1 após envelhecimento
print(f"Idade de p1 => {p1.idade}")
```

## Q&A

Como criar uma classe que consiga mapear todos os elementos (instâncias) desta mesma classe, assim como todos os elementos de classes filhas (derivadas) desta classe parental?



### [Solução](https://stackoverflow.com/questions/12101958/how-to-keep-track-of-class-instances)

sobre-escrever o método de instanciação (criação) da classe desejada.

```python

class BaseClassWithMappingCapabilities(object):
    @classmethod
    def __new__(cls, *args, **kwargs):
        instance = object.__new__(cls)
        if "instances" not in cls.__dict__:
            cls.instances = list()
        if instance not in cls.instances:
          cls.instances.append(instance)

        return instance

class ChildClass(BaseClassWithMappingCapabilities):
  def __init__(self, nome: str) -> None:
      self.nome = nome

  def __repr__(self) -> str:
      return f"Nome: {self.nome}"


classesMapeadoras = [ChildClass(f"Instância {i}") for i in range(1, 11)]

c1 = classesMapeadoras[0]

for sibling in c1.instances:
    print(sibling)
```
