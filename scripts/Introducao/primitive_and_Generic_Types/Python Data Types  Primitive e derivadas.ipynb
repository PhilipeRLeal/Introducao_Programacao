## 2) **Non-Primitive** Type:
    
Tipos não primitivos são tipos de variáveis derivados de tipos primitivos


---



---


  Examplos:

  * **string** (**non-primitive** de **char**)

  * **Tuples**: coleção de dados **imutáveis** de outros tipos **primitivos**. Uma **Tuple** pode ser composta por instâncias **heterogêneas** (não é recomendado) ou  **homogeneas** (recomendado).

  * **list**: coleção de dados **mutáveis** de outros tipos quaisquer (**primitivos**, ou **derivados**). Uma **list** pode ser composta por instâncias **heterogêneas** (não é recomendado) ou  **homogeneas** (recomendado); **list** é armazenado em espaços não sequenciais de memória.

  * **array**: coleção de dados **ordenados**, **homo-tipadas** (*single-typed*) e **mutáveis**; **array** é armazenado em espaços **sequenciais** de memória.

  * **sets**: conjunto de elementos/instâncias **únicas** (sem repetição).. Um **sets** pode ser composto por instâncias **heterogêneas** (não é recomendado) ou  **homogeneas** (recomendado).

  * **dict** (também chamado de **hash** em outras linguagens); serve de dicionário que relaciona uma **chave** (única) a um **valor** (qualquer).

## Arrays

Arrrays are the base blocks for mathematical/statistical studies.

Em Python, arrays advém da biblioteca **numpy**, compilada em C++, para uma máxima otimização do processamento.



```python

import numpy as np

Array = np.array([1,2,3,4,5,6])


Array2 = np.array([2,2,2,2,2,2])


print('Matricial operations:    ', Array * Array2)


# Realizando uma operação de Broadcasting:

print('Broadcasting operations: ', Array * 2)



# Slicing data
print('Slicing: ', Array[:3])



# Transpose operations:
Array = np.arange(9).reshape(3,3)
print('original: \n', Array)
print('transpose  : \n', Array.T)

```

## Sets

```python
x_set = set('CAKE&COKE')

print('Meu set: ', x_set)

y_set = set('COOKIE')

#All elements in x_set that are not in y_set

print('Exclusion of y over x: ', x_set - y_set)

print('Exclusion of x over y: ', y_set - x_set)

# Union
print('Union: ', x_set|y_set)

# Intersection

print('Intersection: ', x_set & y_set)
```

# Objetos Complexos (OOP)

## **Custom classes**:


### Examplo: Sub 20 Futeball team :

```python
class Sub20FutballTeam(dict):
  def __setitem__(self, key, idade:int):
    if idade > 20:
      raise TypeError('{0} could not be put into the team. This person is {1} years old'.format(key, idade))
    else:
      return dict.__setitem__(self, key, idade)

  def show_all_players(self) -> None:
    """Imprime nome e idade de cada jogador no time"""

    print("\n", "Time:")
    for key, value in self.items():
        print(f"\t{key}: {value} anos")

    print()

Sub20 = Sub20FutballTeam()

Sub20['Joao'] = 18

Sub20['Gabriel'] = 19

Sub20.show_all_players()

# print(Sub20)
```
